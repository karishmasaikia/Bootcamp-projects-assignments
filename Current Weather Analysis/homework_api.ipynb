```python
import requests
from config import api_key
import numpy as np
from citipy import citipy
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
#generating the random latitudes and longitudes

lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lons = np.random.uniform(low=-180.000, high=180.000, size=1500)

coordinates=[]
cities_pointer=[]
cities=[]

#generating the coordinates if needed in future
for i in range(1500):
    coordinate_pair=(lats[i],lons[i])
    coordinates.append(coordinate_pair)

#it generates only the pointers if we see by printing cities    
for coordinate_pair in coordinates:
    lat,lon=coordinate_pair
    cities_pointer.append(citipy.nearest_city(lat, lon))
    
#for cities in the list
for city in cities_pointer:
    name = city.city_name
    cities.append(name)

#to remove duplicacies
city_unique=[]

for city in cities:
    if city not in city_unique:
        city_unique.append(city)
    
print(len(cities))  
print(len(city_unique))
```

```python
base_url = "http://api.openweathermap.org/data/2.5/weather?"
unit="imperial"
Total_weather_jsondata=[]
Humidity=[]
Cloudiness=[]
Windspeed=[]
Date=[]
Country=[]
Latitude=[]
Longitude=[]
Max_Temp=[]

#Perform a weather check on each of the cities using a series of successive API calls. 
#Include a print log of each city as it's being processed with the city number, city name, and requested URL.

for city in city_unique:
    query_url = base_url + "appid=" + api_key + "&q=" + city + "&units=" + unit
    print(query_url)
    weather_response=requests.get(query_url)
    weather_jsondata=weather_response.json()
    
#for all the cities, all the parameters were not present when checked the count of rows. so had to do (try, except) for
#all the parameters

    try:
        Max_Temp.append(weather_jsondata['main']['temp_max'])
    except Exception as err:
        Max_Temp.append("NA")
  
    try:
        Humidity.append(weather_jsondata['main']['humidity'])
    except Exception as err:
        Humidity.append("NA")
        
    try:
        Cloudiness.append(weather_jsondata['clouds']['all'])
    except Exception as err:
        Cloudiness.append("NA")
        
    try:
        Windspeed.append(weather_jsondata['wind']['speed'])
    except Exception as err:
        Windspeed.append("NA")
        
    try:
        Date.append(weather_jsondata['dt'])
    except Exception as err:
        Date.append("NA")
        
    try:
        Country.append(weather_jsondata['sys']['country'])
    except Exception as err:
        Country.append("NA")
        
    try:
        Latitude.append(weather_jsondata['coord']['lat'])
        Longitude.append(weather_jsondata['coord']['lon'])
    except Exception as err:
        Latitude.append("NA")
        Longitude.append("NA")

        
        
        
    Total_weather_jsondata.append(weather_jsondata)
    
```

```python
#Changing the date format. Since the date remains same, so Date[0] is taken
Date_format=[]
for i in range(len(Date)):
    Date_format.append(datetime.fromtimestamp(Date[0]).strftime('%Y-%m-%d'))   
```

```python
#Make the dataframe
City_weather_dict={"Cities":city_unique,
                "Cloudiness (%)":Cloudiness,
                "Country":Country,
                "Date":Date_format,
                "Humidity (%)":Humidity,
                "Latitude":Latitude,
                "Longitude":Longitude,
                "Max temp (F)":Max_Temp,
                "Wind Speed (mph)":Windspeed}
    
City_weather_df=pd.DataFrame(City_weather_dict)
City_weather_df.to_csv('weather_details.csv')
City_weather_df.head(3)
```

```python
#removing NA fields from latitude&max temp parameters for plotting purpose
NA_removed_xaxis=City_weather_df.loc[City_weather_df['Latitude']!='NA']
NA_removed_xaxis['Latitude']=pd.to_numeric(NA_removed_xaxis['Latitude'])

NA_removed_xyaxis=NA_removed_xaxis.loc[(NA_removed_xaxis['Max temp (F)'])!='NA']
NA_removed_xyaxis['Max temp (F)']=pd.to_numeric(NA_removed_xyaxis['Max temp (F)'])


plt.figure(figsize=(10,8))
# c is that value in which u want to bring gradient effect.here, for higher values of y values, color gets darkened
#_r=reverse color effect.see color reference for more options
plt.scatter(x=NA_removed_xyaxis["Latitude"],y=NA_removed_xyaxis["Max temp (F)"],c=NA_removed_xyaxis["Max temp (F)"],
            s=75, cmap="hot_r")
            
plt.grid()
plt.xlabel("Latitude",fontsize=15)
plt.ylabel("Maximum Temperature (F)", fontsize=15)
plt.title(f"City Latitude Vs Max Temperature ({Date_format[0]})", fontsize=15)
plt.savefig("Lat_vs_MaxTemp.png")
```

```python
#Humidity (%) vs. Latitude
NA_removed_xyaxis=NA_removed_xaxis.loc[(NA_removed_xaxis['Humidity (%)'])!='NA']
NA_removed_xyaxis['Humidity (%)']=pd.to_numeric(NA_removed_xyaxis['Humidity (%)'])

plt.figure(figsize=(10,8))
plt.scatter(x=NA_removed_xyaxis["Latitude"],y=NA_removed_xyaxis["Humidity (%)"],c=NA_removed_xyaxis["Humidity (%)"],
            cmap="Wistia",s=75)
            
plt.grid()
plt.xlabel("Latitude",fontsize=15)
plt.ylabel("Humidity (%)", fontsize=15)
plt.title(f"City Latitude Vs Humidity (%) ({Date_format[0]})", fontsize=15)
plt.savefig("Lat_vs_Humidity.png")
```

```python
#Cloudiness (%) vs. Latitude
NA_removed_xyaxis=NA_removed_xaxis.loc[(NA_removed_xaxis['Cloudiness (%)'])!='NA']
NA_removed_xyaxis['Cloudiness (%)']=pd.to_numeric(NA_removed_xyaxis['Cloudiness (%)'])

plt.figure(figsize=(10,8))
plt.scatter(x=NA_removed_xyaxis["Latitude"],y=NA_removed_xyaxis["Cloudiness (%)"],s=75)
            
plt.grid()
plt.xlabel("Latitude",fontsize=15)
plt.ylabel("Cloudiness (%)", fontsize=15)
plt.title(f"City Latitude Vs Cloudiness (%) ({Date_format[0]})", fontsize=15)
plt.savefig("Lat_vs_Cloudiness.png")
```

```python
#Wind Speed (mph) vs. Latitude
NA_removed_xyaxis=NA_removed_xaxis.loc[(NA_removed_xaxis['Wind Speed (mph)'])!='NA']
NA_removed_xyaxis['Wind Speed (mph)']=pd.to_numeric(NA_removed_xyaxis['Wind Speed (mph)'])

plt.figure(figsize=(10,8))
plt.scatter(x=NA_removed_xyaxis["Latitude"],y=NA_removed_xyaxis["Wind Speed (mph)"],c=NA_removed_xyaxis["Wind Speed (mph)"],
           s=75,cmap="BuPu_r")
            
plt.grid()
plt.xlabel("Latitude",fontsize=15)
plt.ylabel("Wind Speed (mph)", fontsize=15)
plt.title(f"City Latitude Vs Wind Speed (mph) ({Date_format[0]})", fontsize=15)
plt.savefig("Lat_vs_Windspeed.png")
```
